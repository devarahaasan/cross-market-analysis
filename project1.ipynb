```python
import requests
import pandas as pd
import datetime
import time
```

```python
# step1
all_records=[]# all the datas are into the list
for i in range(1,6):  #Use string formating and for loop and scroll through 5pages to collect data
  url=f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&page={i}&sparkline=False"
  response=requests.get(url) # request to get URL and collect the datas
  if response.status_code==200: # 200 means your request was success
      data=response.json()  # convert data to JSON format
      all_records.extend(data)  #extend data to all_records
```

```python
len(all_records) # check the length
```

```python
records=[] # appended datas are into the list
for i in all_records: # iterate the data from all_records and extract the columns
  records.append(dict(id=i["id"],symbol=i["symbol"],name=i["name"],currentprice=i["current_price"],marketcap=i["market_cap"],marketcaprank=i["market_cap_rank"],totalvolume=i["total_volume"],circulatingsupply=i["circulating_supply"],totalsupply=i["total_supply"],ath=i["ath"],atl=i["atl"],lastupdated=i["last_updated"]))
```

**QUERYS OF TABLE(ONE)**

```python
#Find the top 3 cryptocurrencies by market cap
q1="""
SELECT
       id,name,symbol,marketcap
FROM
   one
ORDER BY
   marketcap DESC LIMIT 3;"""
cursor.execute(q1)
result1=cursor.fetchall()
result1

# define table headers
headers =["id","name","symbol","marketcap"]
# print the result as a table
print(tabulate(result1, headers=headers, tablefmt="grid"))
```

```python
#List all coins where circulating supply exceeds 90% of total supply.
q2="""
SELECT
    id, name,  symbol, circulatingsupply,  totalsupply,
    (circulatingsupply * 100.0 / NULLIF(totalsupply, 0)) as supply_percentage
FROM
  one
WHERE
   (circulatingsupply * 1.0 / NULLIF(totalsupply, 0)) > 0.9
ORDER BY
   supply_percentage DESC;"""

cursor.execute(q2)
result2=cursor.fetchall()
result2

# define table headers
headers=["id","name","symbol","circulatingsupply","totalsupply","supply_percentage"]
#print the result as table
print(tabulate(result2,headers,tablefmt="grid"))
```

```python
#Get coins that are within 10% of their all-time-high (ATH).
q3= """
SELECT
    name,
    currentprice,
    ath,
    ((ath - currentprice) / ath) * 100 AS percentage_ath
FROM
    one
WHERE
    currentprice >= (ath * 0.90)
ORDER BY
    percentage_ath ASC;"""
cursor.execute(q3)
result3=cursor.fetchall()
result3

# define table headers
headers=["name","currentprice","ath","percentage_ath"]
#print the result as table
print(tabulate(result3,headers,tablefmt="grid"))

```

```python
#Find the average market cap rank of coins with volume above $1B
q4="""
SELECT
    AVG(marketcaprank) AS average_rank
FROM
    one
WHERE
    totalvolume > 1000000000;"""

cursor.execute(q4)
result4=cursor.fetchall()
result4

#define table headers
headers=["avgerage_rank"]
#print the result as table
print(tabulate(result4,headers,tablefmt="grid"))
```

```python
#Get the most recently updated coin.
q5="""
SELECT
    id,symbol,name,
    date as recently_updated_coin
FROM
    one
ORDER BY
    date DESC;"""
cursor.execute(q5)
result5=cursor.fetchall()
result5

#define table headers
headers=["id","symbol","name","recently_updated_coin"]
#print the result as table
print(tabulate(result5,headers,tablefmt="grid"))
```

```python
#Compare Bitcoin vs Oil average price in 2025.
j1="""
SELECT
    AVG(b.price_inr) AS avg_bitcoin_price,
    AVG(o.Price_inr) AS avg_oil_price
FROM
   two  as b
INNER JOIN
   three as o  ON b.date = o.Date
WHERE
    b.date BETWEEN '2025-01-01' AND '2025-12-31'
    AND coin_id = 'bitcoin' ;"""
cursor.execute(j1)
result1=cursor.fetchall()
result1

#define table headers
headers=["avg_bitcoin_price","avg_oil_price"]
#print the result as table
print(tabulate(result1,headers,tablefmt="grid"))
```

```python
#Check if Bitcoin moves with S&P 500 (correlation idea).
j2="""
SELECT
    (COUNT(*) * SUM(b.price_inr * s.volume) - SUM(b.price_inr) * SUM(s.volume)) /
    (
      SQRT(COUNT(*) * SUM(b.price_inr * b.price_inr) - SUM(b.price_inr) * SUM(b.price_inr)) *
      SQRT(COUNT(*) * SUM(s.volume * s.volume) - SUM(s.volume) * SUM(s.volume))
    ) AS correlation_coefficient
FROM
    two b
INNER JOIN
    four s ON b.date = s.date
WHERE
    ticker ='^GSPC' ;"""
cursor.execute(j2)
result2=cursor.fetchall()
result2

#define table headers
headers=[" correlation_coefficient"]
#print the result as table
print(tabulate(result2,headers,tablefmt="grid"))
```
